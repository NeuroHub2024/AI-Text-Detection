```python
human_generated_text = data[data.label == 0]
ai_generated_text = data[data.label == 1]
```

```python
human_generated_text
```

```python
ai_generated_text
```

```python
extra_col = human_generated_text.shape[0] - ai_generated_text.shape[0]
extra_col
```

```python
df_dropped = human_generated_text.sample(ai_generated_text.shape[0])
df_dropped
```

```python
df_dropped.reset_index(inplace=True)
df_dropped
```

```python
trimmed_human_genrated_text = df_dropped
trimmed_human_genrated_text
```

```python
new_data = pd.concat([trimmed_human_genrated_text, ai_generated_text])
new_data
```

```python
plt.hist(new_data.label)
```

```python
new_data.reset_index(inplace=True)
```

```python
new_data.drop(['level_0' , 'index'] , axis=1 , inplace=True)
```

```python
new_data
```

```python
def custom(value):
    value = value.replace("\n", "")
#     value = value.replace('\'', "" )
    return value
```

```python
 new_data.text =  new_data.text.apply(custom)
```
